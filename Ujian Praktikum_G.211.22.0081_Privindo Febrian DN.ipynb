**B. Atribut Stock Name dan Symbol yang memiliki Dividen / Share 5 tertinggi**

```python
import matplotlib.pyplot as plt

data = {
    'Stock Name': ['Apple', 'Microsoft', 'Salesforce', 'Oracle', 'Hewlett Packard', 'Alphabet', 'Intel', 'Cisco', 'Qualcomm', 'Amazon', 'Redhat', 'Facebook', 'Twitter'],
    'Symbol': ['AAPL', 'MSFT', 'CRM', 'ORCL', 'HPE', 'GOOG', 'INTC', 'CSCO', 'QCOM', 'AMZN', 'RHT', 'FB', 'TWTR'],
    'Dividen/Share': [2.28, 1.56, 0.00, 0.64, 0.26, 0.00, 1.09, 1.16, 2.12, 0.00, 0.00, 0.00, 0.00]
}

# Mengurutkan data berdasarkan Dividen/Share secara menurun
sorted_data = sorted(zip(data['Stock Name'], data['Symbol'], data['Dividen/Share']), key=lambda x: x[2], reverse=True)

top_5 = sorted_data[:5]  # Mengambil 5 data teratas

stock_names = [item[0] for item in top_5]
symbols = [item[1] for item in top_5]
dividends = [item[2] for item in top_5]

# Membuat bar plot
plt.bar(stock_names, dividends)
plt.xlabel('Stock')
plt.ylabel('Dividen/Share')
plt.title('Top 5 Stocks with Highest Dividen/Share')

plt.xticks(rotation=45)  # Rotasi label sumbu x untuk memudahkan pembacaan

plt.tight_layout()
plt.show()
```
